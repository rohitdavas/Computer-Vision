```python
import numpy as np
import skimage 
import cv2
import matplotlib.pyplot as plt 
```

```python
def plot(images):
    for image in images:
        plt.imshow(image)
        plt.show()
        
```

```python
skimage.lookfor("pyramid")
```

```python
left_image, right_image, disparity = skimage.data.stereo_motorcycle()

plot([left_image, right_image, disparity])
```

## gaussian pyramind

```python
gauss = skimage.transform.pyramid_gaussian(left_image,max_layer= 5,  multichannel= True)


pyr = []
for image in gauss:
    print(image.shape)
    pyr.append(image)
plot(pyr)
print(len(pyr))
```

# laplacian pyramid

```python
laplace = skimage.transform.pyramid_laplacian(left_image, max_layer=5, multichannel= True)
l_pyr = []

for image in laplace:
    l_pyr.append(image)
plot(l_pyr)
```
